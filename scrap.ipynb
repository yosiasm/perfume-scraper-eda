```python
nose = soup.find_all('div',{'class':'cell small-12 medium-4'})
```

```python
nose_clean = {}
for n in nose:
     a = n.find('a')
     url = a['href']
     name = a.text
     nose_clean[url] = name
```

```python
import time
import random
perfurmer_perfume = {}
counter = 0
for u in nose_clean:
    # if u.strip('/noses/')+'.json' in pfile:
    #     counter+=1
    #     continue
    try:
        url = 'https://www.fragrantica.com' + u
        perfumer = scraper.get(url)
        soup2 = bs(perfumer.content)
        detail = soup2.find('div',{'class':'cell auto text-left grid-x align-middle'}).find_all('p')
        detail = [x.text for x in detail]
        perfumes = soup2.find_all('div',{'class':'cell large-6'})
        perfurmer_perfume = {
            'url' : url,
            'detail': detail,
            'perfumes':[]
        }
        for per in perfumes:
            href = per.find('h3').find('a')['href']
            title = per.find('h3').find('span')['title']
            name = per.find('h3').find('a').text
            review = per.find('div',{'style':'color: #909090; margin: 0.2em'})
            review = review.text if review else None
            perfurmer_perfume['perfumes'].append({
                'title':title,
                'href':href,
                'name':name,
                'review':review
            })
        with open('{}.json'.format(u.strip('/')), 'w') as w:
            json.dump(perfurmer_perfume,w)
        time.sleep(random.random())
    except Exception as e:
        print(e)
        time.sleep(60)
    if random.random() > 0.6:
        time.sleep(2)
        print('rest')
    counter+=1
    print(counter,'/',len(nose_clean))
    
```

```python
pfile = os.listdir('noses')
all_perfumes = []
for u in pfile:
    with open(f'noses/{u}','r') as r:
        doc = json.load(r)
        nose = doc['url'].strip('https://www.fragrantica.com/noses/').strip('.html')
        perfumes = doc['perfumes']
        for perf in perfumes:
            perf['nose'] = nose
            all_perfumes.append(perf)
```

```python
perfumes_sample = []
for year in rank_perfume.value_counts('year').index:
    sample = rank_perfume[rank_perfume.year==year].sort_values('review_clean',ascending=False).head(100).to_dict('records')
    print(year, len(sample))
    perfumes_sample.extend(sample)
df_perfumes_sample = pd.DataFrame(perfumes_sample)
df_perfumes_sample
```

```python
import json
import pandas as pd
with open('perfumes_sample.json','r') as r:
    df_perfumes_sample = json.load(r)
df_perfumes_sample = pd.DataFrame(df_perfumes_sample)
df_perfumes_sample
```

```python
import re
import time
import random

counter = 0
for idx,row in df_perfumes_sample.iterrows():
    try:
        perf_url = f'https://www.fragrantica.com{row.href}'
        perfume_html = scraper.get(perf_url)
        soup_perf = bs(perfume_html.content)

        pyramid = soup_perf.find_all('div',{'id':'pyramid'})[0].text
        pyramid = pyramid.replace(' Notes','-notes').strip('Perfume Pyramid').strip('Vote for Ingredients')
        pyramid = re.findall('[A-Z][^A-Z]*', pyramid)
        pyramid_map = {}
        key = pyramid[0]
        for item in pyramid:
            if '-notes' in item:
                key = item
                continue
            if key not in pyramid_map:
                pyramid_map[key] = []
            item = item.replace('\n','')
            pyramid_map[key].append(item)
        #pyramid_map

        accords = soup_perf.find_all('div',{'class':'accord-bar'})
        accords_detail = [{
            'width':acc['style'].strip(';').split(';')[-1].strip('width: ').strip('%'),
            'opacity':acc['style'].strip(';').split(';')[-1].strip('opacity: ').strip('%'),
            'text':acc.text
        } for acc in accords]
        #accords_detail

        reviews = soup_perf.find_all('div',{'itemprop':'reviewBody'})
        reviews = [rev.text for rev in reviews][:20]
        #reviews

        doc = {
            'href':row.href,
            'year':row.year,
            'review_clean':row.review_clean,
            'pyramid':pyramid_map,
            'accords_detail':accords_detail,
            'reviews':reviews
        }
        filename = row.href.strip('/perfume/').strip('.html').replace('/','_')
        with open(f'perfumes/{filename}.json','w') as w:
            json.dump(doc,w)
    except Exception as e:
        print(e)
        time.sleep(90)
    time.sleep(random.random())
    if random.random() > 0.7:
        time.sleep(2)
    time.sleep(5+random.random())
    counter+=1
    print(counter,'/',len(df_perfumes_sample), row.href)
    
    


    
```
